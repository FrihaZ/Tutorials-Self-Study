```python
import requests  ## requests is a library that allows you to send HTTP requests using Python

response= requests.get("http://api.open-notify.org/astros.json")

response.json()
```

```python
jres= response.json()
```

```python
jres=pd.DataFrame(jres)
jres
```

```python
## to remove the dict values into columns based on their keys
## change it into a list using .values.tolist()
jres2 = pd.DataFrame(jres['people'].values.tolist(), index=jres.index)
jres2
```

```python

## merge both dfs based on their index

jres3=pd.merge(jres, jres2, left_index=True, right_index=True)
```

```python
jres3.drop('people', axis=1)  ## remove the people columns axis 1 means column
```
